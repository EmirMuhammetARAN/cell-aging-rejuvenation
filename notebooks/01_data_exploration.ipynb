```python
import json
```

```python
with open('../data/raw/train/via_project.json') as f:
    data = json.load(f)

senescent_cells = []
young_cells = []

for img_id, img_data in data.items():
    regions = img_data['regions']
    for region in regions:
        region_attributes = region['region_attributes']
        for key, value in region_attributes.items():
            if key == 'cell':
                if value == 'senescent':
                    senescent_cells.append(region)
                elif value == 'young':
                    young_cells.append(region)
print(f"Number of senescent cells: {len(senescent_cells)}")
print(f"Number of young cells: {len(young_cells)}")
```

```python
w_sum, h_sum = 0, 0
for senescent_img in senescent_cells:
    shape_attributes = senescent_img['shape_attributes']
    x, y = shape_attributes['all_points_x'], shape_attributes['all_points_y']
    w,h = max(x)-min(x), max(y)-min(y)
    w_sum += w
    h_sum += h

avg_w = w_sum / len(senescent_cells)
avg_h = h_sum / len(senescent_cells)
print(f"Average width of senescent cells: {avg_w}")
print(f"Average height of senescent cells: {avg_h}")
```

```python
w_sum, h_sum = 0, 0
for young_img in young_cells:
    shape_attributes = young_img['shape_attributes']
    x, y = shape_attributes['all_points_x'], shape_attributes['all_points_y']
    w,h = max(x)-min(x), max(y)-min(y)
    w_sum += w
    h_sum += h

avg_w = w_sum / len(young_cells)
avg_h = h_sum / len(young_cells)
print(f"Average width of young cells: {avg_w}")
print(f"Average height of young cells: {avg_h}")
```
